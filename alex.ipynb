```python
import pandas as pd
import matplotlib
url = "https://raw.githubusercontent.com/GWC-DCMB/CapstoneProject/master/datasets/"
filepath = "tesla/Tesla_Deaths.csv"
df = pd.read_csv(url + filepath).fillna(0)
teslaDeaths = 0
otherDeaths = 0
df.values

for i in range(0,144):
    df.loc[i, 'Non_Occupant'] = df.loc[i,"Other_vehicle"] + df.loc[i,"Cyclists_Peds"]
    otherDeaths += df.loc[i,"Other_vehicle"] + df.loc[i,"Cyclists_Peds"]
    df.loc[i, 'Occupant'] = df.loc[i,"Tesla_occupant"] + df.loc[i,"Tesla_driver"]
    teslaDeaths += df.loc[i,"Tesla_occupant"] + df.loc[i,"Tesla_driver"]
#df[["Deaths","Tesla_driver","Tesla_occupant","Other_vehicle","Cyclists_Peds","Non_Occupant","Occupant"]]

deaths = pd.DataFrame({"label": [""]})
deaths.loc[0,"label"] = "Other Deaths"
deaths.loc[0,"deaths"] = otherDeaths
deaths.loc[1,"label"] = "Tesla Deaths"
deaths.loc[1,"deaths"] = teslaDeaths
# deaths.plot.bar(x="teslaDeaths")
deaths.plot.bar(x="label",y="deaths")
```
